```python
import pandas as pd
data={
    "name":['ram','shyam','ganshyam','sohan','mohan','joe','roy','liger'],
    "age":[22,23,24,25,26,27,28,16],
    "city":['delhi','mumbai','goa','delhi','mumbai','goa','delhi','mumbai'],
    "salary":[10000,20000,30000,40000,50000,60000,70000,80000]

}
df=pd.DataFrame(data)
print("salary is greater than 50000")
high_salary=df[df["salary"]>50000]
print(high_salary)
#filtered salary and age
print("salary is greater than 50000 and age is >21")
filtered_data=df[(df["salary"]>50000) & (df["age"]>21)]
print(filtered_data)
# filtered salary or age
print("salary is greater than 50000 or age is >21")
filtered_data=df[(df["salary"]>50000) | (df["age"]>21)]
print(filtered_data)
```

```python
#handling missing data
import pandas as pd
data={
    "name":['ram',None,'ganshyam','sohan','mohan','joe','roy','liger'],
    "age":[22,23,24,None,26,27,28,16],
    "city":['delhi','mumbai','goa',None,'mumbai','goa','delhi','mumbai'],
    "salary":[10000,20000,30000,None,50000,60000,70000,80000]

}
df=pd.DataFrame(data)

#print(df.isnull().sum())
#df.dropna(axis=0,inplace=True)
df["age"].fillna(df["age"].mean(),inplace=True)
df["salary"].fillna(df["salary"].mean(),inplace=True)
print(df)
```

```python
#aggregation method mean(),max(),min(),count(),sum()
import pandas as pd
data={
    "name":['ram','shyam','ganshyam','sohan','mohan','joe','roy','liger'],
    "age":[22,23,24,25,26,27,28,16],
    "city":['delhi','mumbai','goa','delhi','mumbai','goa','delhi','mumbai'],
    "salary":[10000,20000,30000,40000,50000,60000,70000,80000]

}
df=pd.DataFrame(data)
avg_salary=df['salary'].mean()
print(avg_salary)
max_salary=df['salary'].max()
print(max_salary)
min_salary=df['salary'].min()
print(min_salary)
sum_salary=df['salary'].sum()
print(sum_salary)
count_salary=df['salary'].count()
print(count_salary)
```

```python
import pandas as pd

df_customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['ram', 'shyam', 'ganshyam', 'sohan', 'mohan']
})

df_orders = pd.DataFrame({
    'CustomerID': [1, 2, 3, 7, 8],
    'OrderAmount': [250, 350, 450, 750, 950]
})

df_merged = pd.merge(df_customers, df_orders, on='CustomerID', how='inner')

print('inner join')
print(df_merged)
```

```python
#concatenation
import pandas as pd
df_region1=pd.DataFrame ({
    'CustomerID':[1,2,3,4,5],
    'Name':['ram','shyam','ganshyam','joy','roy']
})
df_region2=pd.DataFrame({
    'CustomerID':[6,7,8,9,10],
    'Name':['liger','joe','mohan','sohan','roe']

})
concatenated_df=pd.concat([df_region1,df_region2],axis=0,ignore_index=True)
print(concatenated_df)
```
